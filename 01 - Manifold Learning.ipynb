### Get datasets

```python
import numpy as np
import pandas as pd
```

```python
def get_spiral(n_samples, n_rolls, noise=0.5, random_state=42) -> tuple[np.ndarray, np.ndarray]:
    "Return the spiral dataset and the univariate position of the sample according to the main dimension of the points in the manifold."
    
    if random_state is not None:
        np.random.seed(random_state)

    # Generate the univariate position
    t = np.linspace(0, 2 * np.pi * n_rolls, n_samples)
    
    # Generate the spiral dataset
    x = t * np.cos(t)
    y = t * np.sin(t)
    
    # Add noise
    x += noise * np.random.randn(n_samples)
    y += noise * np.random.randn(n_samples)
    
    return np.vstack((x, y)).T, t


X, t = get_spiral(300, 3)
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], c=t, s=5);
```

```python
from sklearn.datasets import make_swiss_roll

X, t = make_swiss_roll(n_samples=600, noise=0.5, random_state=42)

# 3D Plot of the dataset
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.scatter(X[:,0], X[:,1], X[:,2], s=5, c=t)

# 2D Plot of the dataset
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,2], s=5, c=t);
```

```python
from sklearn.datasets import make_s_curve

X, t = make_s_curve(n_samples=1000, noise=0.1, random_state=42)

# 3D Plot of the dataset
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.scatter(X[:,0], X[:,1], X[:,2], s=5, c=t)

# 2D Plot of the dataset
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,2], s=5, c=t);
```

### Isomap
Implement the Isomap algorithm using scikit-learn and visualize the results.

```python
from sklearn.manifold import Isomap

X, t = make_swiss_roll(n_samples=1500, noise=0.5, random_state=42)

# Isomap
model = Isomap(n_components=2, n_neighbors=5)
X_iso = model.fit_transform(X)

# 2D Plot of the dataset
fig, ax = plt.subplots()
ax.scatter(X_iso[:,0], X_iso[:,1], s=5, c=t);
```

### Locally Linear Embedding (LLE)
Implement the LLE algorithm using scikit-learn and visualize the results.

```python
from sklearn.manifold import LocallyLinearEmbedding

X, t = make_s_curve(n_samples=1500, noise=0.1, random_state=42)

# LLE
model = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_lle = model.fit_transform(X)

# 2D Plot of the dataset
fig, ax = plt.subplots()
ax.scatter(X_lle[:,0], X_lle[:,1], s=5, c=t);
```
